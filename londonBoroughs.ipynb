Best practice is for pandas DataFrames to contain the **observations of interest** as rows, and the features of those observations as columns. You want **tidy** DataFrames: whose rows are observations and whose columns are variables.


Notice here that the column headings are the *particular* boroughs, which is your observation of interest. Most of the other cell-values are the average proprety values of the borough corresponding to that time stamp. 

Clearly, you need to roll up your sleeves and do some  **cleaning**.

You've made some progress! But with new progress comes new issues. For one, the row indices of our DataFrame contain the names of the boroughs. You should never have a piece of information we want to analyze as an index, this information should be within the DataFrame itself. The indices should just be a unique ID, almost always a number.

Those names are perhaps the most important piece of information! Put them where you can work with them.

```python
dfg = df.groupby(by=['London_Borough', 'Year']).mean()
dfg.sample(10)
# groupby() fonksiyonu ile fiyatlarin orta degerlerini bulduk
```

```python
dfg = dfg.reset_index()
dfg.head()
# Indekslerimizi resetledik
```

### 3. Modelling


```python
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
# d argumani alan bir fonksiyon olusturduk bu fonksiyon 1998 ve 2018 icin ayri ayri ortalama fiyat degerlerini bulup bunlarin oranini hesapliyor
```

```python
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
final = {}
# Bu isi butun ilceler icin yapip modelleme yapacagimiz icin bir sozluk olusturduk
```

```python
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough'] == b]
    final[b] = create_price_ratio(borough)
print(final)
# Bir for dongusu yardimiyle her ilce icin create_price_ratio fonksiyonunu calistirip final sozlugune kaydettik
```

```python
df_ratios = pd.DataFrame(final)
# Pandas kutuphanesindeki DataFrame fonksiyonu yardimiyla sozlugumuzu DataFrame'e cevirdik
```

```python
df_ratios.head()
# Kontrol
```

```python
df_ratios_T = df_ratios.T
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
# Yukarida yaptigimiz gibi transpozunu alip indeks resetledik
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
# Sutunlarin isimlerini duzelttik
```

```python
top15 = df_ratios.sort_values(by='2018',ascending=False).head(15)
print(top15)
# Azalan seklinde siralayip ilk 15 ilceyi top15 degiskenine kaydettik
```

```python
ax = top15[['Borough','2018']].plot(kind='bar')

ax.set_xticklabels(top15.Borough)
# plot() yardimiyla bar seklindeki grafigimizi olusturduk
```
